```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.datasets import load_iris

# 階層型クラスタリングの実施
# ウォード法 x ユークリッド距離
linkage_result = linkage(df, method='ward', metric='euclidean')

# クラスタ分けするしきい値を決める
#threshold = 0.5 * np.max(linkage_result[:, 2])

# 階層型クラスタリングの可視化
plt.figure(num=None, figsize=(16, 9), dpi=200, facecolor='w', edgecolor='k')
dendrogram(linkage_result,
           orientation="left",
           labels=df.index, 
           color_threshold=55)
plt.axvline(x=55, c='grey', lw=1, linestyle='dashed')
plt.savefig('Python_dendrogram2.png', dpi=150)
```
